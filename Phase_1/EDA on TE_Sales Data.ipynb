```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
testData = pd.read_csv('TE Sales Labelled Data- Analysis testData.csv')
```

```python
testData.head()
```

```python
relaventColumns = [
    "abicms_caseid", "abicms_casemanagementid", "abicms_casemanagementidname", 
    "abicms_casenumber", "abicms_casetransactionsid", "abicms_categorytes", 
    "abicms_comments", "abicms_subcategorytes", "abicms_expectationofriskscorename",
    "abicms_typeofrisktename", "modifiedby", "modifiedbyname", "createdby", 
    "createdbyname", "createdonbehalfby", "createdonbehalfbyname", 
    "abicms_typeofrisktesalesname", "abicms_riskscore"
]
```

```python
testData_Relavent = testData[relaventColumns]
```

```python
testData_Relavent.isnull().sum()
```

```python
testData_Relavent.drop(columns=["createdonbehalfby", "createdonbehalfbyname", "abicms_categorytes", "abicms_comments", "abicms_subcategorytes"])
```

```python
testData_Relavent = testData_Relavent[relaventColumns].copy()
testData_Relavent["abicms_riskscore"].fillna(testData_Relavent["abicms_riskscore"].mean(), inplace=True)
```

```python
numericaDataInsights = testData_Relavent.describe()
categoricalDataInsights = testData_Relavent.select_dtypes(include=['object']).nunique()
mfvCategorical = testData_Relavent.select_dtypes(include=['object']).mode().iloc[0]

riskScoreInsights = {
    "Min Risk Score": testData_Relavent["abicms_riskscore"].min(),
    "Max Risk Score": testData_Relavent["abicms_riskscore"].max(),
    "Mean Risk Score": testData_Relavent["abicms_riskscore"].mean(),
    "Median Risk Score": testData_Relavent["abicms_riskscore"].median(),
    "Standard Deviation": testData_Relavent["abicms_riskscore"].std()
}

caseCreationCounts = testData_Relavent["createdbyname"].value_counts()
caseModificationCounts = testData_Relavent["modifiedbyname"].value_counts()

riskExpectationCounts = testData_Relavent["abicms_expectationofriskscorename"].value_counts()

riskAvgScore = testData_Relavent.groupby("abicms_expectationofriskscorename")["abicms_riskscore"].mean()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(testData_Relavent["abicms_riskscore"], bins=10, kde=True, color="skyblue")
plt.axvline(testData_Relavent["abicms_riskscore"].mean(), color='red', linestyle='dashed', linewidth=1, label="Mean")
plt.title("Distribution of Risk Scores")
plt.xlabel("Risk Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#Case Counts
created_by_counts = testData_Relavent["createdbyname"].value_counts()
modified_by_counts = testData_Relavent["modifiedbyname"].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
#Creation
sns.barplot(y=created_by_counts.index, x=created_by_counts.values, ax=axes[0], palette="Blues_r")
axes[0].set_title("Number of Cases Created by Users")
axes[0].set_xlabel("Number of Cases")
axes[0].set_ylabel("Created By")

#Modification
sns.barplot(y=modified_by_counts.index, x=modified_by_counts.values, ax=axes[1], palette="Greens_r")
axes[1].set_title("Number of Cases Modified by Users")
axes[1].set_xlabel("Number of Cases")
axes[1].set_ylabel("Modified By")

plt.tight_layout()
plt.show()
```

```python
risk_by_creator = testData_Relavent.groupby("createdbyname")["abicms_riskscore"].mean().sort_values()
plt.figure(figsize=(10, 5))
sns.barplot(y=risk_by_creator.index, x=risk_by_creator.values, palette="coolwarm")
plt.title("Average Risk Score by Case Creator")
plt.xlabel("Average Risk Score")
plt.ylabel("Created By")
plt.show()
```

```python
risk_expectation_counts = testData_Relavent["abicms_expectationofriskscorename"].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(y=risk_expectation_counts.index, x=risk_expectation_counts.values, palette="Purples_r")
plt.title("Distribution of Risk Expectation Levels")
plt.xlabel("Number of Cases")
plt.ylabel("Risk Expectation Level")
plt.show()
```

```python
risk_expectation_avg_score = testData_Relavent.groupby("abicms_expectationofriskscorename")["abicms_riskscore"].mean().sort_values()
plt.figure(figsize=(8, 5))
sns.barplot(y=risk_expectation_avg_score.index, x=risk_expectation_avg_score.values, palette="Oranges_r")
plt.title("Average Risk Score by Expectation Level")
plt.xlabel("Average Risk Score")
plt.ylabel("Risk Expectation Level")
plt.show()
```
